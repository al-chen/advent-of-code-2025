```python
from pathlib import Path

data = Path("../inputs/day01.txt").read_text()
rotations = [(line[0], int(line[1:])) for line in data.splitlines()]

N = 100
START = 50
```

```python
# part 1
position = START
count = 1 if position == 0 else 0

for direction, clicks in rotations:
    position += clicks if direction == 'R' else -clicks
    position %= N
    if position == 0:
        count += 1

count
```

```python
# part 2
position = START
count = 1 if position == 0 else 0

for direction, clicks in rotations:
    count += clicks // N

    clicks %= N
    if clicks == 0:
        continue

    prev = position
    position += clicks if direction == 'R' else -clicks

    if position < 0:
        position += N
        if prev > 0:
            count += 1
    elif position == 0 or position == N:
        position = 0
        count += 1
    elif position > N - 1:
        position -= N
        count += 1

count
```
