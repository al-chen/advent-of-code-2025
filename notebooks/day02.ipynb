```python
from pathlib import Path

data = Path("../inputs/day02.txt").read_text()
ranges = [r.split('-') for r in data.split(',')]
```

```python
# part 1
def is_invalid(digits):
    half = len(digits) // 2
    return digits[:half] == digits[half:]

def invalid_sum(low, high, predicate):
    return sum(id for id in range(int(low), int(high) + 1) if predicate(str(id)))

sum(invalid_sum(low, high, is_invalid) for low, high in ranges)
```

```python
# part 2
def is_repeating(digits, pattern_len):
    pattern = digits[:pattern_len]
    return len(digits) % pattern_len == 0 and digits == pattern * (len(digits) // pattern_len)

def is_invalid_part2(digits):
    return any(is_repeating(digits, length) for length in range(1, len(digits) // 2 + 1))

sum(invalid_sum(low, high, is_invalid_part2) for low, high in ranges)
```
