```python
from pathlib import Path

data = Path("../inputs/day03.txt").read_text()
banks = data.splitlines()
```

```python
# part 1

def extract_largest_digits(line, count):
    """Extract the `count` largest digits from line, preserving left-to-right order."""
    digits = []
    remaining = line
    for i in range(count):
        searchable = remaining[:len(remaining) - (count - i - 1)]
        largest = max(searchable)
        idx = remaining.index(largest)
        digits.append(largest)
        remaining = remaining[idx + 1:]
    return int("".join(digits))

total = sum(extract_largest_digits(line, 2) for line in banks)
print(total)
```

```python
# part 2

total = sum(extract_largest_digits(line, 12) for line in banks)
print(total)
```
